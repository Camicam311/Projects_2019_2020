```python
class Perceptron():
    
    def __init__(self, n =1000, lr = 1.0001):
        self.w = None
        self.bias = None
        self.lr = lr
        self.n = n    
        self.model = self._unit_step_func
    

    def fit(self, X, Y):
        n_samples, n_features = X.shape       
        # init parameters
        self.w = np.zeros(n_features)
        self.bias = 0
        # number of iterations required
        for _ in range(self.n):
            # in the sequence (x1, y1), ..., (xt, yt)
            for i in range(n_samples):
                # if yt<wt, xt> <= 0
                x = X.iloc[i, :]
                y = Y[i]
                output = np.dot(x, self.w) + self.bias
                y_pred = self.model(output)
                update = self.lr * (y - y_pred)
                self.w += x * update
                self.bias += update
                    
    def _unit_step_func(self, sum):
        return np.where(sum >= 0.0, 1, 0)
    
    def predict(self, X):
        output = np.dot(X, self.w) + self.bias
        return self.model(output)
    
    def error(self, Y_pred, Y):
        return np.mean(Y_pred != Y)
    
    def gen_conf_matx(self, t_x, t_y):
        conf_matx = pd.DataFrame(0, index=range(1,8), columns=range(1,7))

        for i in range(len(t_y)):
            conf_matx.loc[int(t_x[i]), int(t_y[i])] += 1

        return round(conf_matx/conf_matx.sum(), 3)              
```

```python
class VPerceptron():
    
    def __init__(self, n =1000, lr = 1):
        self.w = None
        self.bias = None
        self.c = None
        self.m = None
        self.lr = lr
        self.n = n    
        self.model = self._unit_step_func

    def fit(self, X, Y):
        n_samples, n_features = X.shape       
        # init parameters
        self.bias = 0
        self.m = 0
        self.w = [0]
        self.c = [1]
        # number of iterations required
        for _ in range(self.n):
            # in the sequence (x1, y1), ..., (xt, yt)
            for i in range(n_samples):
                x = X.iloc[i, :]
                y = Y[i]
                w = self.w[-1]
                c = self.c[-1]
                output = np.dot(x, w) + self.bias
                y_pred = self.model(output)
                update = self.lr * (y - y_pred)
                     
                # if Y_t<W_m, X_t> <= 0
                if update.all() != 0:
                    #W_m+1 = W_m + Y_t * X_t
                    new_m = w + x * update 
                    self.w.append(new_m)
                    # m = m + 1
                    self.m += 1
                    self.bias += c * update
                    # C_m = 1
                    self.c.append(1)
                else:     
                    self.c[-1] += 1
                    
    def _unit_step_func(self, sum):
        return np.where(sum >= 0.0, 1, 0)
     
    def predict(self, X):
        outputs = np.zeros(X.shape[0])
        for i in range(self.m):
            
            # sign<wi, x>
            output = np.dot(X, self.w[i]) + self.bias[0]
            if i == 0:
                output = [x[0] for x in output]
                output = np.asarray(output)
                
            # ci * sign<wi, x>
            y_pred = self.c[i] * self.model(output)
            outputs += y_pred
        threshold = np.sum(self.c)/2.0
        
        # sign(Eci*sign<wi, x>)
        return np.where(outputs >= threshold, 1, 0)
    
    def error(self, Y_pred, Y):
        return np.mean(Y_pred != Y)
```

```python
vperceptron = VPerceptron(n = 1)
wt_matrix = vperceptron.fit(X_train_, Y_train_)
Y_pred_ = vperceptron.predict(X_train_)
error = vperceptron.error(Y_pred_, Y_train_)
error
```

```python
class APerceptron():
    
    def __init__(self, n =1000, lr = 1):
        self.w = None
        self.bias = None
        self.c = None
        self.lr = lr
        self.n = n  
        self.m = None
        self.model = self._unit_step_func

    def fit(self, X, Y):
        n_samples, n_features = X.shape       
        # init parameters
        self.bias = 0
        self.m = 0
        self.w = [0]
        self.c = [1]
        # number of iterations required
        for _ in range(self.n):
            # in the sequence (x1, y1), ..., (xt, yt)
            for i in range(n_samples):
                x = X.iloc[i, :]
                y = Y[i]
                w = self.w[-1]
                c = self.c[-1]
                output = np.dot(x, c * w) + self.bias
                y_pred = self.model(output)
                update = self.lr * (y - y_pred)
                     
                # if Y_t<W_m, X_t> <= 0
                if update.all() != 0:
                    #W_m+1 = W_m + Y_t * X_t
                    new_m = w + x * update 
                    self.w.append(new_m)
                    # m = m + 1
                    self.m += 1
                    self.bias += update
                    # C_m = 1
                    self.c.append(1)
                else:     
                    self.c[-1] += 1
                    
    def _unit_step_func(self, sum):
        return np.where(sum >= 0, 1, 0)
     
    def predict(self, X):
        product = 0
        for i in range(self.m):
            product += (self.c[i] * self.w[i] + self.bias)
        output = np.dot(X, product) 
        
        return self.model(output) 
    
    def error(self, Y_pred, Y):
        return np.mean(Y_pred != Y)
```

```python
class APerceptron_coordinate():
    
    def __init__(self, n =1000, lr = 1):
        self.w = None
        self.bias = None
        self.c = None
        self.lr = lr
        self.n = n    
        self.model = self._unit_step_func

    def fit(self, X, Y):
        n_samples, n_features = X.shape       
        # init parameters
        self.bias = 0
        self.m = 0
        self.w = [0]
        self.c = [1]
        # number of iterations required
        for _ in range(self.n):
            # in the sequence (x1, y1), ..., (xt, yt)
            for i in range(n_samples):
                x = X.iloc[i, :]
                y = Y[i]
                w = self.w[-1]
                c = self.c[-1]
                output = np.dot(x, c * w) + self.bias
                y_pred = self.model(output)
                update = self.lr * (y - y_pred)
                     
                # if Y_t<W_m, X_t> <= 0
                if update.all() != 0:
                    #W_m+1 = W_m + Y_t * X_t
                    new_m = w + x * update 
                    self.w.append(new_m)
                    # m = m + 1
                    self.m += 1
                    self.bias += update
                    # C_m = 1
                    self.c.append(1)
                else:     
                    self.c[-1] += 1
        w_avg = np.mean([w * c for w, c in 
                        zip(self.w, self.c)])
        return w_avg
                    
    def _unit_step_func(self, sum):
        return np.where(sum >= 0, 1, 0)
```

```python
highest = []
lowest = []
for i in (2, 3, 4):
    perceptron = APerceptron_coordinate(n = i)
    w = perceptron.fit(X_train_, Y_train_).tolist()
    series = pd.Series(w, index=columns[:-1])
    sort = series.sort_values()
    least = sort[:4]
    most = sort[-4:]
    highest.append(most)
    lowest.append(least)
highest, lowest
```

```python
y = pd.DataFrame()
for i in np.arange(1, 7):
    train_i = train.copy()
    train_x = train_i.iloc[:, :-1]
    train_y = train_i.iloc[:, -1].map(lambda x: 
                                      1 if x == i else 0)
    c_i = Perceptron(n=1)
    c_i.fit(train_x, train_y)
    
    test_i = test.copy()
    test_x = test_i.iloc[:, :-1]
    test_y = test_i.iloc[:, -1]
    
    pred_y = c_i.predict(test_x)
    pred_y = pd.Series(pred_y).map({1: i, 0: 7})
    y[i] = pred_y
y['actl'] = test.iloc[:, -1]
y.head()
```

```python
matx = pd.DataFrame(0, index=range(1,8), columns=range(1,7))
for i in range(len(y)):
    ys_ = y.iloc[i, :].tolist()
    actl = y.iloc[i, -1]
    ys = [y for y in ys_ if y != 7.0]
    if len(ys) == 2:
        # prediction y is correct
        if ys[0] == ys[1]:
            matx.loc[actl, actl] += 1
        # prediction y and aactual y are different
        else:
            matx.loc[ys[0], actl] += 1
    else:
        pred = ys[:-1]
        mode = stats.mode(pred)
        # predictions y correct
        if np.mean([y == int(actl) for y in ys]) == 1.0:
            matx.loc[actl, actl] += 1
        # predictions y incorrect
        elif np.mean([y == pred[-1] for y in pred]) == 1.0:
            matx.loc[pred[-1], actl] += 1
        # predictions y ambiguous, "don't know"
        else:
            matx.loc[7, actl] += 1                
confusion_matrix = round(matx/matx.sum(), 3)
confusion_matrix 
```
